```python
# Custom loss function

percentiles = torch.arange(0, 100.0001, 1, dtype=torch.float32) / 100.0
#print(percentiles)

# Loss MADc Squared
def MADcLossSquared(zTrain_tensor, Y_pred):
    target_prc = torch.quantile(zTrain_tensor, percentiles, dim=0)
    pred_prc = torch.quantile(Y_pred, percentiles, dim=0)
    mad_prc = torch.mean(torch.abs(target_prc - pred_prc))
    loss = torch.nanmean(torch.abs(zTrain_tensor - Y_pred)) + 2*mad_prc
    return loss ** 2  # Square the loss
        
```

```python
# MLR and MLP models

# Linear regression model
class LinearRegression(nn.Module):
    def __init__(self, input_size):
        super(LinearRegression, self).__init__()
        self.linear = nn.Linear(input_size, 1)

    def forward(self, x):
        return self.linear(x)


hiddenLayerSizes = [120, 120]
class mlpModel(torch.nn.Module):
    def __init__(self, device, inputSize, outputSize):
        super(mlpModel, self).__init__()

        self.inputLr = torch.nn.Linear(inputSize, hiddenLayerSizes[0], device=device)

        self.hiddenLayers = torch.nn.ModuleList()  # ✅ IMPORTANT FIX
        for ihl in range(len(hiddenLayerSizes)-1):
            self.hiddenLayers.append(
                torch.nn.Linear(hiddenLayerSizes[ihl], hiddenLayerSizes[ihl+1], device=device)
            )

        self.outputLr = torch.nn.Linear(hiddenLayerSizes[-1], outputSize, device=device)
        self.relu = torch.nn.ReLU()

    def forward(self, x):
        out = self.relu(self.inputLr(x))
        for hl in self.hiddenLayers:
            out = self.relu(hl(out))
        out = self.outputLr(out)
        return out
```

```python
from scipy.signal import find_peaks

# Unified Permutation Importance (GPU, vectorized)
# ================================================
def permutation_importance_gpu(
    model, X, y, metric=torch.mean, n_repeats=10, device='cuda',
    mode="global", percentile=99
):
    """
    Compute permutation importance for a model on GPU.

    Args:
        model: Trained PyTorch model
        X (torch.Tensor): Input features [n_samples, n_features]
        y (torch.Tensor): Targets [n_samples]
        metric: Metric function (default = torch.mean for MAE)
        n_repeats (int): Number of permutations
        device (str): 'cuda' or 'cpu'
        mode (str): "global" for all samples, "peaks" for surge peaks above percentile
        percentile (float): Percentile threshold for peak selection (used if mode="peaks")
    
    Returns:
        np.ndarray: Permutation importance scores for each feature
    """
    model.eval()
    y = y.view(-1)

    # Select subset of samples depending on mode
    if mode == "peaks":
        # Work on CPU first for peak detection
        y_np = y.detach().cpu().numpy()
        threshold = np.percentile(y_np, percentile)
        peaks, _ = find_peaks(y_np, height=threshold)

        if len(peaks) == 0:
            raise ValueError(f"No peaks found above the {percentile}th percentile.")

        X = X[peaks]
        y = y[peaks]

    n_samples, n_features = X.shape

    with torch.no_grad():
        baseline_pred = model(X).squeeze()
        baseline_score = metric(torch.abs(y - baseline_pred))

    importances = []

    for i in range(n_features):
        # Create n_repeats permuted versions of X
        X_repeat = X.unsqueeze(0).repeat(n_repeats, 1, 1)
        for r in range(n_repeats):
            perm_indices = torch.randperm(n_samples, device=device)
            X_repeat[r, :, i] = X_repeat[r, perm_indices, i]

        # Flatten to (n_repeats * n_samples, n_features)
        X_batch = X_repeat.view(-1, n_features)

        with torch.no_grad():
            pred_batch = model(X_batch).squeeze()

        # Reshape back to (n_repeats, n_samples)
        pred_batch = pred_batch.view(n_repeats, n_samples)

        # Metric per repeat
        scores = metric(torch.abs(pred_batch - y.unsqueeze(0)), dim=1)
        scores = scores - baseline_score

        importances.append(scores.mean().item())

    return np.array(importances)
```

```python
# loop over saved checkpoints

import os

inputSize = XTest_tensor.shape[1]   # number of features (should be 29)
outputSize = 1

num_runs = 40

for run in range(1, num_runs + 1):

    checkpoint_filename = f"mlpModel_Run{run}_checkpoint.pth"
    base_filename = os.path.splitext(checkpoint_filename)[0]

    # Initialize model
    model = mlpModel(device, inputSize, outputSize).to(device)

    # Load weights
    model.load_state_dict(torch.load(checkpoint_filename, map_location=device))
    model.eval()

    # GLOBAL importance
    importances_global = permutation_importance_gpu(
        model,
        XTest_tensor.to(device),
        ZTest_tensor.to(device).squeeze(),
        n_repeats=10,
        device=device,
        mode="global"
    )

    np.savetxt(
        f"{base_filename}_permutation_importance_global.txt",
        importances_global,
        fmt="%.6f"
    )

    # PEAKS importance
    importances_peaks = permutation_importance_gpu(
        model,
        XTest_tensor.to(device),
        ZTest_tensor.to(device).squeeze(),
        n_repeats=10,
        device=device,
        mode="peaks",
        percentile=99
    )

    np.savetxt(
        f"{base_filename}_permutation_importance_peaks.txt",
        importances_peaks,
        fmt="%.6f"
    )

    print(f"Run {run}: permutation importance computed.")
```

```python
# Permutation importance plot considering all runs - Full time series

import glob

ml_method = 'MLP-MSE'
#ml_method = r'$\mathrm{MLP-MADc}^2$'

# === Custom feature names ===
feature_names = [
    "SSH-PC1", "SSH-PC2", "SSH-PC3", "SSH-PC4", "SSH-PC5", "SSH-PC6", "SSH-PC7",
    "Tide",
    "MSLP-PC1", "MSLP-PC2", "MSLP-PC3", "MSLP-PC4", "MSLP-PC5", "MSLP-PC6", "MSLP-PC7",
    "WSx-PC1", "WSx-PC2", "WSx-PC3", "WSx-PC4", "WSx-PC5", "WSx-PC6", "WSx-PC7",
    "WSy-PC1", "WSy-PC2", "WSy-PC3", "WSy-PC4", "WSy-PC5", "WSy-PC6", "WSy-PC7"
]

# === Define color map for groups ===
group_colors = {
    "SSH": "royalblue",
    "Tide": "gray",
    "MSLP": "darkorange",
    "WSx": "seagreen",
    "WSy": "purple"
}

def get_color(feature):
    if feature.startswith("SSH"):
        return group_colors["SSH"]
    elif feature == "Tide":
        return group_colors["Tide"]
    elif feature.startswith("MSLP"):
        return group_colors["MSLP"]
    elif feature.startswith("WSx"):
        return group_colors["WSx"]
    elif feature.startswith("WSy"):
        return group_colors["WSy"]
    else:
        return "black"  # fallback

# === Load all permutation importance results ===
files = sorted(glob.glob("*_permutation_importance_global.txt"))
importances = [np.loadtxt(file) for file in files]
importances = np.vstack(importances)  # shape (num_runs, num_features)

# === Compute mean and std across runs ===
mean_importance = importances.mean(axis=0)
std_importance = importances.std(axis=0)

# === Sort features by importance (descending) ===
sorted_idx = np.argsort(mean_importance)[::-1]
sorted_features = [feature_names[i] for i in sorted_idx]
sorted_mean = mean_importance[sorted_idx]
sorted_std = std_importance[sorted_idx]
sorted_colors = [get_color(f) for f in sorted_features]

# === Plot ===
plt.rcParams['font.family'] = 'Century Gothic'  # Replace with your preferred font
# Ensure MathText uses the same font
plt.rcParams['mathtext.default'] = 'rm'  # Use the regular font for MathText
plt.rcParams['mathtext.fontset'] = 'custom'  # Enable custom font for MathText
plt.rcParams['mathtext.rm'] = 'Century Gothic'  # Set Roman font for MathText

plt.figure(figsize=(14, 6))
bars = plt.bar(range(len(sorted_mean)), sorted_mean, yerr=sorted_std, capsize=5,
               color=sorted_colors, edgecolor="black")

plt.xticks(range(len(sorted_mean)), sorted_features, rotation=90)
plt.ylabel("Δ error", fontsize=18)
plt.title(f'Trieste - Permutation importance full time series: {ml_method}', fontsize=22)

plt.tick_params(axis='both', labelsize=18)
plt.ylim(-0.003, 0.051)

# === Legend for groups ===
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor=group_colors["SSH"], edgecolor="black", label="SSH"),
    Patch(facecolor=group_colors["Tide"], edgecolor="black", label="Tide"),
    Patch(facecolor=group_colors["MSLP"], edgecolor="black", label="MSLP"),
    Patch(facecolor=group_colors["WSx"], edgecolor="black", label="WSx"),
    Patch(facecolor=group_colors["WSy"], edgecolor="black", label="WSy")
]
plt.legend(handles=legend_elements, loc="upper right", fontsize=14)

plt.tight_layout()
#plt.show()

# Clean the method name for the filename
ml_method_clean = (
    ml_method.replace('$', '')  # Remove LaTeX math symbols
    .replace(r'\mathrm{', '')  # Remove the \mathrm{...}
    .replace('}', '')          # Remove closing braces
    .replace('{', '')          # Remove opening braces
    .replace('^', '')          # Remove caret for superscripts
    .replace('-', '_')         # Replace dash with underscore
)
# Generate the filename
filename = f'PI_targetTG_Trieste_{ml_method_clean}.png'
# Save the figure with transparent axes but white background
plt.savefig(filename, dpi=300, bbox_inches='tight')
```
