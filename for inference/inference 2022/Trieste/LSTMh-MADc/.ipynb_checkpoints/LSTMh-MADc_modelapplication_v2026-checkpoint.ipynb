```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

import pandas as pd 

import numpy as np

from scipy.signal import detrend
from sklearn.preprocessing import StandardScaler
from scipy.stats import pearsonr
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import TheilSenRegressor
from scipy.stats import ecdf

import matplotlib.pyplot as plt

import sympy

import time
import joblib

# Load predictors data

cmems_path = 'C:/Users/Rodrigo/Desktop/Rodrigo/02 - Research fellow/01 - Storm surge ML downscaling/ML downscaling november 2022/Predictors/cmems_7PCs_fulldomain_Trieste_2022.txt'
tide_path = 'C:/Users/Rodrigo/Desktop/Rodrigo/02 - Research fellow/01 - Storm surge ML downscaling/ML downscaling november 2022/Predictors/tide_Trieste_2022.txt'
mslp_path = 'C:/Users/Rodrigo/Desktop/Rodrigo/02 - Research fellow/01 - Storm surge ML downscaling/ML downscaling november 2022/Predictors/ERA5mslp_7PCs_fulldomain_Trieste_2022.txt'
u10_path = 'C:/Users/Rodrigo/Desktop/Rodrigo/02 - Research fellow/01 - Storm surge ML downscaling/ML downscaling november 2022/Predictors/ERA5stressX_7PCs_fulldomain_Trieste_2022.txt'
v10_path = 'C:/Users/Rodrigo/Desktop/Rodrigo/02 - Research fellow/01 - Storm surge ML downscaling/ML downscaling november 2022/Predictors/ERA5stressY_7PCs_fulldomain_Trieste_2022.txt'

cmems = pd.read_csv(cmems_path,header=None,sep=r"\s+")
tide = pd.read_csv(tide_path,header=None,sep=r"\s+")
mslp = pd.read_csv(mslp_path,header=None,sep=r"\s+")
u10 = pd.read_csv(u10_path,header=None,sep=r"\s+")
v10 = pd.read_csv(v10_path,header=None,sep=r"\s+")


# Data filter on years of interest
anio = 2022
offset = 1  # En Python el conteo parte desde cero

# Predictors
idini_cmems = cmems.index[cmems.iloc[:, 0] == anio].tolist()
idini_mslp = mslp.index[mslp.iloc[:, 0] == anio].tolist()
idini_tide = tide.index[tide.iloc[:, 0] == anio].tolist()

idfin_cmems = cmems.index[cmems.iloc[:, 0] == anio].tolist()
idfin_mslp = mslp.index[mslp.iloc[:, 0] == anio].tolist()
idfin_tide = tide.index[tide.iloc[:, 0] == anio].tolist()

# Ensure there are valid indices before slicing
if idini_cmems and idfin_cmems:
    cmemsTest = cmems.iloc[idini_cmems[offset]:idfin_cmems[-1] + 1, :]
if idini_mslp and idfin_mslp:
    mslpTest = mslp.iloc[idini_mslp[offset]:idfin_mslp[-1] + 1, :]
    u10Test = u10.iloc[idini_mslp[offset]:idfin_mslp[-1] + 1, :]
    v10Test = v10.iloc[idini_mslp[offset]:idfin_mslp[-1] + 1, :]
if idini_tide and idfin_tide:
    tideTest = tide.iloc[idini_tide[offset]:idfin_tide[-1] + 1, :]


# Data for statistical downscaling (Testing data)

x_prd = cmemsTest.iloc[:, 7:14] #- cmemsTest.iloc[:, 7].mean()
#x_prd = cmemsTest.iloc[:, 7] - cmemsTest.iloc[:, 7].mean() / cmemsTest.iloc[:, 7].std()
y_prd = tideTest.iloc[:, 7] #- tideTest.iloc[:, 7].mean()
#y_prd = tideTest.iloc[:, 7] - tideTest.iloc[:, 7].mean() / tideTest.iloc[:, 7].std()
w_prd = mslpTest.iloc[:, 7:14]
#w_prd = mslpTest.iloc[:, 7] - mslpTest.iloc[:, 7].mean() / mslpTest.iloc[:, 7].std()
u_prd = u10Test.iloc[:, 7:14]
#u_prd = u10Test.iloc[:, 7] - u10Test.iloc[:, 7].mean() / u10Test.iloc[:, 7].std()
v_prd = v10Test.iloc[:, 7:14]
#v_prd = v10Test.iloc[:, 7] - v10Test.iloc[:, 7].mean() / v10Test.iloc[:, 7].std()

# Detrend sea level series
x1_prd_dtn = detrend(x_prd[7])
x2_prd_dtn = detrend(x_prd[8])
x3_prd_dtn = detrend(x_prd[9])
x4_prd_dtn = detrend(x_prd[10])
x5_prd_dtn = detrend(x_prd[11])
x6_prd_dtn = detrend(x_prd[12])
x7_prd_dtn = detrend(x_prd[13])

# Creating XTest DataFrame and standarization
XTest = pd.DataFrame({'x1_dtn': x1_prd_dtn, 'x2_dtn': x2_prd_dtn, 'x3_dtn': x3_prd_dtn, 'x4_dtn': x4_prd_dtn, 'x5_dtn': x5_prd_dtn, 'x6_dtn': x6_prd_dtn, 'x7_dtn': x7_prd_dtn,\
                      'y': y_prd,\
                      'w1': w_prd[7], 'w2': w_prd[8], 'w3': w_prd[9], 'w4': w_prd[10], 'w5': w_prd[11], 'w6': w_prd[12], 'w7': w_prd[13],\
                      'u1': u_prd[7], 'u2': u_prd[8], 'u3': u_prd[9], 'u4': u_prd[10], 'u5': u_prd[11], 'u6': u_prd[12], 'u7': u_prd[13],\
                      'v1': v_prd[7], 'v2': v_prd[8], 'v3': v_prd[9], 'v4': v_prd[10], 'v5': v_prd[11], 'v6': v_prd[12], 'v7': v_prd[13]})

# Load preprocessing metadata (training scaler)
preproc = joblib.load(
    "C:/Users/Rodrigo/Desktop/Rodrigo/02 - Research fellow/01 - Storm surge ML downscaling/Experiments_v6_v2/surge-ml_v2 - sliding-window/experiments/Trieste/LSTMh-MADc/lstmHybridModel_Run34_preprocessing.pkl"
)

scaler = preproc["scaler"]
feature_names_train = preproc["feature_names"]

XTest = XTest[feature_names_train]

# Apply training scaler (no fitting!)
XTest = pd.DataFrame(
    scaler.transform(XTest),
    columns=feature_names_train
)

print("\n--- CORRECT SCALING (training scaler) ---")

print("Overall mean (scaled):", np.mean(XTest.values))
print("Overall std  (scaled):", np.std(XTest.values))

print("\nPer-feature mean:")
print(XTest.mean().values)

print("\nPer-feature std:")
print(XTest.std().values)

XTest = np.clip(XTest, -8, 8)
# Convert data to PyTorch tensors
#Test data
XTest_np = XTest.to_numpy()            # Convert DataFrame to NumPy array
XTest_tensor = torch.tensor(XTest_np, dtype=torch.float32)     # Convert NumPy array to PyTorch tensor


# Custom loss function

percentiles = torch.arange(0, 100.0001, 1, dtype=torch.float32) / 100.0
#percentiles2 = torch.arange(99, 100.0000001, .01, dtype=torch.float32) / 100.0
#percentiles = torch.hstack([percentiles, percentiles2])
#print(percentiles)

# Loss MADc Squared
def MADcLossSquared(zTrain_tensor, Y_pred):
    target_prc = torch.quantile(zTrain_tensor, percentiles, dim=0)
    pred_prc = torch.quantile(Y_pred, percentiles, dim=0)
    mad_prc = torch.mean(torch.abs(target_prc - pred_prc))
    #loss = torch.nanmean(torch.abs(zTrain_tensor - Y_pred)) + 2*mad_prc
    loss = torch.nanmean(torch.abs(zTrain_tensor - Y_pred)) + mad_prc
    return loss ** 2  # Square the loss
        

# Define the modelsclass:

class lstmHybridModel(torch.nn.Module):
    def __init__(self, input_size, output_size, hidden_layer_size):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_layer_size, num_layers=1, batch_first=True)
        self.lin = nn.Linear(input_size, hidden_layer_size)
        self.relu = nn.ReLU()
        self.output = nn.Linear(2 * hidden_layer_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.lstm.num_layers, x.size(0), self.lstm.hidden_size, device=x.device)
        c0 = torch.zeros(self.lstm.num_layers, x.size(0), self.lstm.hidden_size, device=x.device)
        out_rnn, _ = self.lstm(x, (h0, c0))
        out_lin = self.relu(self.lin(x))[:, -1, :]
        out = torch.cat((out_lin, out_rnn[:, -1, :]), dim=1)
        return self.output(out)

# Load checkpoint
checkpoint = torch.load("lstmHybridModel_Run34_checkpoint.pth", map_location="cpu")

# Print checkpoint keys
print(checkpoint.keys())

# If it's a state_dict, print its keys
if isinstance(checkpoint, dict):
    if "state_dict" in checkpoint:
        print(checkpoint["state_dict"].keys())  # Check stored model keys
    else:
        print(checkpoint.keys())  # Print available keys


# Define device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Set random seed for reproducibility
seed_value = 0
torch.manual_seed(seed_value)  
torch.cuda.manual_seed_all(seed_value)

# Disable non-deterministic algorithms on CUDA
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
torch.use_deterministic_algorithms(False)  # Ensure full determinism

# Load checkpoint 
checkpoint = torch.load("lstmHybridModel_Run34_checkpoint.pth", map_location=device, weights_only=True)

# Initialize LSTM model
input_size = XTest.shape[1]  # Make sure this matches your training input_size
output_size = 1  # Single output prediction

# Create model instance
model = lstmHybridModel(input_size=input_size, output_size=1, hidden_layer_size=60)

# Load the weights from the checkpoint
model.load_state_dict(checkpoint, strict=False)  # Allow missing keys if necessary

# Move model to device
model.to(device)

# Prepare input tensor
predictors = XTest_tensor.to(device)  # Ensure it's on the correct device

# Reshape for LSTM: (seq_len, batch_size, input_size)
predictors = predictors.unsqueeze(1)  # If needed, adjust for LSTM input format

# Set evaluation mode
model.eval()

# Make predictions
with torch.no_grad():
    predictions = model(predictors)  # Ensure proper sequence handling

# Convert predictions to a numpy array for plotting
predictions2 = predictions.cpu().detach().numpy()

# Compute basic statistics
mean_pred = np.mean(predictions2)
max_pred = np.max(predictions2)
min_pred = np.min(predictions2)

# Print results
print(f"Mean Prediction: {mean_pred:.4f}")
print(f"Max Prediction: {max_pred:.4f}")
print(f"Min Prediction: {min_pred:.4f}")

# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(predictions2.flatten(), label='Predictions', color='blue', linestyle='-')
plt.title('Model Predictions')
plt.xlabel('Sample Index')
plt.ylabel('Predicted Value')
plt.legend()
plt.show()

# Reset index of test data (if needed)
mslpTest_reset_index = mslpTest.reset_index(drop=True)

# Concatenate test data with predictions2
m = pd.concat([mslpTest_reset_index.iloc[:, :7], pd.DataFrame(predictions2)], axis=1)

# Save to a .txt file
np.savetxt("lstmHybridModel_Trieste_2022_checkpoint34_v2026.txt", m.values, fmt='%f', delimiter=' ')
```
