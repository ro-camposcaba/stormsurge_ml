```python
# Convert data to PyTorch tensors

#Test data
XTest_np = XTest.to_numpy()            # Convert DataFrame to NumPy array
XTest_tensor = torch.tensor(XTest_np, dtype=torch.float32)     # Convert NumPy array to PyTorch tensor
```

```python
# Custom loss function

percentiles = torch.arange(0, 100.0001, 1, dtype=torch.float32) / 100.0
#percentiles2 = torch.arange(99, 100.0000001, .01, dtype=torch.float32) / 100.0
#percentiles = torch.hstack([percentiles, percentiles2])
#print(percentiles)

# Loss MADc Squared
def MADcLossSquared(zTrain_tensor, Y_pred):
    target_prc = torch.quantile(zTrain_tensor, percentiles, dim=0)
    pred_prc = torch.quantile(Y_pred, percentiles, dim=0)
    mad_prc = torch.mean(torch.abs(target_prc - pred_prc))
    #loss = torch.nanmean(torch.abs(zTrain_tensor - Y_pred)) + 2*mad_prc
    loss = torch.nanmean(torch.abs(zTrain_tensor - Y_pred)) + mad_prc
    return loss ** 2  # Square the loss
        
```

```python
# Define the models: Multivariate Linear Regression and Multilayer Perceptron

# Linear regression model
class LinearRegression(nn.Module):
    def __init__(self, input_size):
        super(LinearRegression, self).__init__()
        self.linear = nn.Linear(input_size, 1)

    def forward(self, x):
        return self.linear(x)

hiddenLayerSizes = [120, 120] 

class mlpModel(torch.nn.Module):
    def __init__(self, device, inputSize, outputSize):
        super(mlpModel, self).__init__()
        self.inputSize = inputSize
        self.outputSize = outputSize

        self.inputLr = torch.nn.Linear(self.inputSize, hiddenLayerSizes[0], device=device)

        # Use ModuleList instead of a Python list
        self.hiddenLayers = torch.nn.ModuleList()
        for ihl in range(len(hiddenLayerSizes)-1):
            hlInputSize = hiddenLayerSizes[ihl]
            hlOutputSize = hiddenLayerSizes[ihl + 1]
            hl = torch.nn.Linear(hlInputSize, hlOutputSize, device=device)
            self.hiddenLayers.append(hl)

        self.outputLr = torch.nn.Linear(hiddenLayerSizes[-1], self.outputSize, device=device)
        self.relu = torch.nn.ReLU()

    def forward(self, x):
        out = self.inputLr(x)
        out = self.relu(out)
        for hlr in self.hiddenLayers:  # Iterate over ModuleList
            out = hlr(out)
            out = self.relu(out)
        out = self.outputLr(out)
        return out
```

```python
# Define device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Set random seed for reproducibility
seed_value = 0
torch.manual_seed(seed_value)  
torch.cuda.manual_seed_all(seed_value)

# Disable non-deterministic algorithms on CUDA
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
torch.use_deterministic_algorithms(False)  # Ensure full determinism

# Load checkpoint for the linear regression model
checkpoint = torch.load("LinearRegression_MADc_Trieste_TG_Run17_checkpoint.pth", map_location=device, weights_only=True)

# Initialize Linear Regression model
input_size = XTest.shape[1]  # Ensure this matches the input features of XTest
model = LinearRegression(input_size)

# Load model state dictionary
model.load_state_dict(checkpoint)

# Move model to the appropriate device
model.to(device)

# Convert test data to tensor if needed and move to device
predictors = XTest_tensor.to(device)

# Set model to evaluation mode
model.eval()

# Make predictions
with torch.no_grad():
    predictions = model(predictors)

# Convert predictions to numpy array for visualization
predictions2 = predictions.cpu().detach().numpy()

# Compute basic statistics
mean_pred = np.mean(predictions2)
max_pred = np.max(predictions2)
min_pred = np.min(predictions2)

# Print results
print(f"Mean Prediction: {mean_pred:.4f}")
print(f"Max Prediction: {max_pred:.4f}")
print(f"Min Prediction: {min_pred:.4f}")

# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(predictions2, label='Linear Regression Predictions', color='blue', linestyle='-')
plt.title('Linear Regression Model Predictions')
plt.xlabel('Sample Index')
plt.ylabel('Predicted Value')
plt.legend()
plt.show()
```
