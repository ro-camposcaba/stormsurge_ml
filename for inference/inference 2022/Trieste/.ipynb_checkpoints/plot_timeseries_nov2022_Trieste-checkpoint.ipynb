```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# --------------------------------------------------
# File paths
# --------------------------------------------------

obs_file = r'C:\Users\Rodrigo\Desktop\Rodrigo\02 - Research fellow\01 - Storm surge ML downscaling\Experiments_v6_v2\surge-ml_v2 - sliding-window\downscaling 2022\observations\meteoDTNLP13h_TG_Trieste_2022.txt'

cop_file = r'C:\Users\Rodrigo\Desktop\Rodrigo\02 - Research fellow\01 - Storm surge ML downscaling\Experiments_v6_v2\surge-ml_v2 - sliding-window\downscaling 2022\medphysics\meteoDTNLP13h_cmemsforecast_Trieste_2022-2023.txt'

mlr_file = r'C:\Users\Rodrigo\Desktop\Rodrigo\02 - Research fellow\01 - Storm surge ML downscaling\Experiments_v6_v2\surge-ml_v2 - sliding-window\downscaling 2022\Trieste\MLR-MADc\mlrModelLP13h_Trieste_2022_checkpoint26_v2026.txt'

lstm_file = r'C:\Users\Rodrigo\Desktop\Rodrigo\02 - Research fellow\01 - Storm surge ML downscaling\Experiments_v6_v2\surge-ml_v2 - sliding-window\downscaling 2022\Trieste\LSTMh-MADc\lstmHybridModelLP13h_Trieste_2022_checkpoint34_v2026.txt'


# --------------------------------------------------
# Load data
# --------------------------------------------------

sta = pd.read_csv(obs_file, delim_whitespace=True, header=None)
cop = pd.read_csv(cop_file, delim_whitespace=True, header=None)
mlr = pd.read_csv(mlr_file, delim_whitespace=True, header=None)
lstm = pd.read_csv(lstm_file, delim_whitespace=True, header=None)


# --------------------------------------------------
# Build datetime column
# (columns: year, month, day, hour, minute, second)
# --------------------------------------------------

def build_time(df):
    return pd.to_datetime(dict(
        year=df[0],
        month=df[1],
        day=df[2],
        hour=df[3],
        minute=df[4],
        second=df[5]
    ))

sta["time"] = build_time(sta)
cop["time"] = build_time(cop)
mlr["time"] = build_time(mlr)
lstm["time"] = build_time(lstm)


# --------------------------------------------------
# Remove mean (surge anomaly)
# --------------------------------------------------

x = sta[7] - sta[7].mean()
f = cop[7] - cop[7].mean()
y1 = mlr[7] - mlr[7].mean()
y2 = lstm[7] - lstm[7].mean()


# --------------------------------------------------
# Plot full downscaling period
# --------------------------------------------------

plt.figure(figsize=(17,5))

plt.plot(sta["time"], x, 'k', lw=1.4, label="Tide gauge: Trieste")
plt.plot(cop["time"], f, 'm', lw=1.4, label="med-Physics")
plt.plot(mlr["time"], y1, color=[0,0.4470,0.7410], lw=1.2, label="MLR-MADc$^2$")
plt.plot(lstm["time"], y2, color=[0.8500,0.3250,0.0980], lw=1.2, label="LSTMh-MADc$^2$")

plt.title("Trieste, 2022", fontname="Century Gothic", fontweight="bold")
plt.xlabel("Time [mm/yy]", fontname="Century Gothic")
plt.ylabel("Surge [m]", fontname="Century Gothic")

plt.grid(True)

ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter("%m/%y"))
ax.tick_params(labelsize=16)

plt.legend(loc="upper center", ncol=4)

plt.tight_layout()

# plt.savefig("Timeseries_2022_traintestTG_Trieste_MLR_LSTMh_HQ.png", dpi=400)


# --------------------------------------------------
# Specific storm surge event (November 2022)
# --------------------------------------------------

start = "2022-11-01"
end = "2022-11-30 23:00"

plt.figure(figsize=(18,6.5))

plt.plot(sta["time"], x, 'k', lw=1.8, label="TG: Trieste")
plt.plot(cop["time"], f, 'm', lw=1.4, label="Med-Physics")
plt.plot(mlr["time"], y1, color=[0,0.4470,0.7410], lw=1.6, label="MLR-MADc$^2$")
plt.plot(lstm["time"], y2, color=[0.8500,0.3250,0.0980], lw=1.6, label="LSTMh-MADc$^2$")

plt.xlim(pd.Timestamp(start), pd.Timestamp(end))
plt.ylim([-0.3,1.1])

plt.title("Trieste (November, 2022)", fontname="Century Gothic", fontweight="bold", fontsize = 20)
plt.xlabel("Time [dd/mm/yy]", fontname="Century Gothic", fontsize = 20)
plt.ylabel("Surge [m]", fontname="Century Gothic", fontsize = 20)

ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter("%d/%m/%y"))
ax.tick_params(labelsize=20)

plt.grid(True)

plt.legend(loc="lower center", bbox_to_anchor=(0.5,-0.25), ncol=4)

plt.tight_layout()

# plt.savefig("Timeseries_november2022_traintestTG_Trieste_MLR_LSTMh_HQ.png", dpi=400)

plt.show()
```
