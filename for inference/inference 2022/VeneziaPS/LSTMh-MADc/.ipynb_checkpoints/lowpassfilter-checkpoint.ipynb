```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
import datetime

# Load data
data = np.loadtxt('lstmHybridModel_VeneziaPS_2022_checkpoint15.txt')
t = data[:, 6]  # Adjusted for 0-based index
x = data[:, 7]

# Sampling parameters
dt = 3600  # Sampling time in seconds
fs = 1 / dt  # Sampling frequency in Hz
cop = 13 * 3600  # Cut-off period in seconds
cof = 1 / cop  # Cut-off frequency

# Design low-pass FIR filter
d = signal.firwin(numtaps=9, cutoff=cof, fs=fs, pass_zero=True)

# Handle NaNs by filtering only valid data
y = np.full_like(x, np.nan)  # Initialize y with NaNs
valid_mask = ~np.isnan(x)  # Mask valid values
y[valid_mask] = signal.filtfilt(d, 1, x[valid_mask])  # Filter only valid data

# Plot results
plt.figure()
plt.plot(t, x, label='Original')
plt.plot(t, y, label='Low-pass filtered', linestyle='dashed')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Signal')
plt.grid()
plt.show()

# Save filtered data while preserving NaNs
met = np.column_stack((data[:, :7], y))
np.savetxt('lstmHybridModelLP13h_VeneziaPS_2022_checkpoint15.txt', met, fmt='%.6f', delimiter=' ')
```
