```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
from scipy.stats import gaussian_kde
from scipy.signal import find_peaks
import seaborn as sns

# Load best run data 
shy = np.loadtxt('lstmHybridModel_Run33_testresults.txt')

lim_inf = -1
lim_sup = 1.2

# Remove the mean
xx = shy[:, 8] - np.nanmean(shy[:, 8])  # Target (observed)
yy = shy[:, 7] - np.nanmean(shy[:, 7])  # Predicted

# Define storm surge threshold (modify as needed)
#storm_surge_threshold = 0.37  # Example threshold, adjust accordingly
storm_surge_threshold = np.percentile(xx, 99)

# Identify peaks in the target series
peaks, _ = find_peaks(xx, height=storm_surge_threshold)  # Extract peaks above threshold

# Extract peak values for target (x) and corresponding predicted values (y)
x = xx[peaks]
y = yy[peaks]

# Debugging checks
print(f"Number of storm surge peaks: {len(x)}")
print(f"Minimum peak value: {np.min(x) if len(x) > 0 else 'N/A'}")

# Linear Regression
if len(x) > 1:
    x_with_intercept = sm.add_constant(x)  # Adds a constant (intercept) to the input
    model = sm.OLS(y, x_with_intercept).fit()  # Perform linear regression
    coeff = model.params
    Y_md = coeff[1] * x + coeff[0]  # Linear model prediction: Y = mx + b

    # Create values for prediction line (Ysp)
    vv = np.linspace(lim_inf, lim_sup, len(y))  
    Ysp = coeff[1] * vv + coeff[0]

    # Removing NaN values
    idx = np.isnan(x)
    ub = np.where(idx == 0)[0]  # Indices where NaN is not present
    xr = x[ub]
    yr = y[ub]

    # Correlation
    r = np.corrcoef(xr, yr)[0, 1]

    # RMSE
    rmse_test = np.sqrt(np.mean((xr - yr) ** 2))

    # Bias
    bias = np.sum(yr - xr) / len(xr)

    # MAD (Mean Absolute Deviation)
    mad = np.nanmean(np.abs(x - y))

    # Percentiles
    sta_prc = np.percentile(x, np.arange(0, 101, 1))
    shy_prc = np.percentile(y, np.arange(0, 101, 1))
    
    # Other statistics for model evaluation
    mad_prc = np.nanmean(np.abs(shy_prc - sta_prc))
    mad3rd = mad + mad_prc

    # Print results
    print(f'SLF: {round(coeff[1],3)}')
    print(f'Correlation: {round(r,3)}')
    print(f'RMSE: {round(rmse_test,3)}')
    print(f'Bias: {round(bias,3)}')
    print(f'MAD: {round(mad,3)}')
    print(f'MADp: {round(mad_prc,3)}')
    print(f'MADc: {round(mad3rd,3)}')
else:
    print("Not enough storm surge peaks detected for regression analysis.")
```
