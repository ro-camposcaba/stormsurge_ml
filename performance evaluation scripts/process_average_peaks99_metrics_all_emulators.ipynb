```python
import pandas as pd

# =================================================
# FILES
# =================================================
FILE_TRI = "Trieste/Trieste_Error_Metrics_Peaks99_All_Emulators.xlsx"
FILE_VEN = "VeneziaPS/VeneziaPS_Error_Metrics_Peaks99_All_Emulators.xlsx"
OUTPUT_FILE = "Average_Error_Metrics_Peaks99_All_Emulators.xlsx"

# =================================================
# LOAD EXCEL FILES
# =================================================
xls_tri = pd.ExcelFile(FILE_TRI)
xls_ven = pd.ExcelFile(FILE_VEN)

# Sanity check: same emulators
assert set(xls_tri.sheet_names) == set(xls_ven.sheet_names), \
    "Sheet names (emulators) do not match!"

# =================================================
# PROCESS EACH EMULATOR
# =================================================
writer = pd.ExcelWriter(OUTPUT_FILE, engine="openpyxl")

for sheet in xls_tri.sheet_names:

    print(f"Averaging peaks-only results for emulator: {sheet}")

    df_tri = pd.read_excel(xls_tri, sheet_name=sheet, index_col=0)
    df_ven = pd.read_excel(xls_ven, sheet_name=sheet, index_col=0)

    # Ensure same runs and ordering
    df_tri = df_tri.sort_index()
    df_ven = df_ven.sort_index()

    # Average across locations (run-by-run)
    df_avg = (df_tri + df_ven) / 2

    # Save averaged sheet
    df_avg.to_excel(writer, sheet_name=sheet, float_format="%.4f")

writer.close()

print(f"\n✅ Averaged peak-based results saved to: {OUTPUT_FILE}")
```
